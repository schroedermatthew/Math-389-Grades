## **1B)**
A = $\begin{bmatrix} 1 & 0 & -4\\ 0 & 5 & 4\\ -4& 4 & 3 \end{bmatrix}$

(a) Find the eigenvalues and corresponding eigenvectors of ${A}$

(b) Is ${A}$ similar to a diagonal matrix? If so, find a nonsingular matrix ${P}$ such that $P^{-1} AP$ is diagonal. Is ${P}$ unique? Explain?

(c) Find the eigenvalues of ${A^-1}$

(d) Find the eigenvalues and corresponding eignvectors of ${A^2}$

### **answers**
-------------------------
**(a)** $A$ is symmetric, first find the characteristic equation of $A$

$\qquad 0 =$ det$(A - \lambda I) = \begin{bmatrix} 1-\lambda & 0 & -4 \\ 0 & 5-\lambda & 4 \\ -4 & 4 & 3-\lambda \end{bmatrix} = (3-\lambda)(\lambda-9)(\lambda+3)$

eigenvalues:
$\qquad \lambda_1 = 3, \qquad \lambda_2 = 9, \qquad \lambda_3 = -3$

since all are distict $A$ can be diagonalized

For $ \lambda_1 = 3 \Rightarrow$
$\begin{bmatrix} -2 & 0 & -4 \\ 0 & 2 & 4 \\ -4 & 4 & 0 \end{bmatrix} \rightarrow$
$\begin{bmatrix} -2 & 0 & -4 \\ 0 & 2 & 4 \\ 0 & 0 & 0 \end{bmatrix} \rightarrow$
$\begin{bmatrix} 1 & 0 & 2 \\ 0 & 1 & 2 \\ 0 & 0 & 0 \end{bmatrix} \Rightarrow$
$ v_1 = \begin{bmatrix}-2 \\ -2 \\ 1 \end{bmatrix}$


For $ \lambda_2 = 9 \Rightarrow v_2 = \begin{bmatrix}1 \\ -2 \\ -2 \end{bmatrix}$


For $ \lambda_3 = -3 \Rightarrow v_3 = \begin{bmatrix}-2 \\ 1 \\ -2 \end{bmatrix}$

------------------

**(b)** $A$ is symmetric and therfore diagonalized

$ P = (v_1, v_2, v_3) = \begin{bmatrix} -2 & 1 & -2 \\ -2 & -2 & 1 \\ 1 & -2 & -2 \end{bmatrix}$

Then $ P^-1 AP = \begin{bmatrix} 3 & 0 & 0 \\ 0 & 9 & 0 \\ 0 & 0 & -3 \end{bmatrix} = D$

P is not unique since eigenvectors associated with eigenvalues are not unique

----------------------

**(c)** if $A$ is invertible and has eigenvalue $\lambda \neq 0$ then the eigenvalue of $A^{-1}$ is 1 / $\lambda$

So $A^{-1}$'s eigenvlaues are 1/3, 1/9, and -1/3

----------------

**(d)** eigenvalues for $A^2$

$\lambda$ is an eigenvalue of A with eigenvector $\xi$ Then $A^2\xi = A(\lambda\xi) = \lambda A \xi = \lambda^2\xi$

$\lambda^2$ is an eigenvalue of $A^2$ with eigenvector $\xi$ so the eigenvalues of $A^2$ are: $\quad3^2 = 9\quad$ $9^2 = 81\quad$ $-3^2 = 9\quad$
and assoiated eigenvectors $\quad v_1, v_2, v_3$

## **3A)**

Let

$\vec{u} = \begin{bmatrix} 1 \\ 0 \\ 1 \end{bmatrix} 
\qquad 
\vec{v} = \begin{bmatrix} 1 \\ -1 \\ 0 \end{bmatrix} 
\qquad 
\vec{w} = \begin{bmatrix} 2 \\ 2 \\ -\sqrt{6} \end{bmatrix}$

(a) Find ||$\vec{u}$||, ||$\vec{v}$||. Find a unit vector in the direction of $\vec{u}$.

(b) Find the distance between $\vec{v}$ and $\vec{w}$

(c) Find the angle between $\vec{u}$ and $\vec{v}$

(d) Show that $\vec{v}$ and $\vec{w}$ are orthogonal

### **answers**
---------------------
**(a)** ||$u$|| = $\sqrt{1^2 + 0^2 + 1^2} \Rightarrow \sqrt{2}\qquad$ and ||$v$|| = $\sqrt{1^2 + (-1)^2 + 0^2} \Rightarrow \sqrt{2}$

so $\quad \large \hat{u} = \frac{u}{||u||} = \frac{1}{\sqrt{2}} \begin{pmatrix} 1 \\ 0 \\ 1 \end{pmatrix}$

-----------------------------

**(b)** distance between $v$ and $w$

$\quad ||v - w|| = \sqrt{(1-2)^2 + (-1-2)^2 + (0 -(-\sqrt{6})^2} = \sqrt{1 + 9 + 6} \Rightarrow 4$

------------------------

**(c)** angle between $u$ and $v$ denoted by $\theta$

$\quad u \cdot v = ||u||||v|| \cos \theta$

$\quad \cos\theta = \large \frac{u \cdot v}{||u|| ||v||} = \frac{1}{\sqrt{2}\sqrt{2}} = \frac{1}{2}$

$\quad \theta = \large \arccos({\frac{1}{2}}) = \frac{\pi}{3}$

----------------------------

**(d)** show $v$ and $w$ are orthogonal

$\quad v \cdot w = 1 \times 2 +(-1) \times 2 + 0 \times (-\sqrt{6}) = 2 - 2 = 0$

So, $v$ and $w$ are orthogonal

---------------------------

## **3B)** 

(a) Prove the Cauch-Schwarz Inequality: If ${u}$ and ${v}$ are any vectors in an inner product space ${V}$, then $\langle u, v \rangle ^2 \leq ||u||^2 ||v||^2$ 

(b) Consider $\mathbb{R}^n$ with the standard inner product. Let $u = (u_1, u_2, ... , u_n)$ and $ v = (v_1, v_2, ... , v_n)$ prove that 

$\large(\sum_{i=1}^{n}u_i,v_i)^2 = (\sum_{i=1}^{n}u_i^2)(\sum_{i=1}^{n}v_i^2)$

(c) Let $ V $ be the vector space of all continuous real-valued functions on the unit interval [0,1] with the inner product $\langle f, g \rangle = \int_0^1f(t)g(t)dt $. Prove 

$\large(\int_0^1f(t)g(t)df)^2 \leq (\int_0^1f^2 (t)dt)(\int_0^1g^2(t)df)$

## **answers**

----------------------------

**(a)** assume $v \neq 0$ and consider the orthogonal projecion

$\qquad \large z = u - \frac{\langle u, v\rangle}{\langle v, v\rangle} v$

since $z$ and $v$ are orthogonal

$\qquad \large u = z + \frac{\langle u, v\rangle}{\langle v, v\rangle} v$

then 

$\qquad \large ||u||^2 = ||z||^2 + \frac{{\langle u, v\rangle}^2}{{\langle v, v\rangle}^2} ||v||^2 = ||z||^2 + \frac{{\langle u, v\rangle}^2}{||v||^2} \geq \frac{{\langle u, v\rangle}^2}{||v||^2}$

So, 

$\qquad \large \langle u, v \rangle ^2 \leq ||u||^2 ||v||^2$

------------------------------

**(b)** 

$\qquad \large (\sum_{i=1}^{n} u_i v_i)^2 = \langle u, v \rangle$

and

$\qquad \large \sum_{i=1}^{n} u_i^2 = \langle u, u \rangle = ||u||^2$

and

$\qquad \large \sum_{i=1}^{n} v_i^2 = \langle v, v \rangle = ||v||^2$

So by Cauchy-Schwartz Inequality the inequality holds

------------------------------

**(c)**

$\qquad \large (\int_{0}^{1}f(t)g(t)dt)^2 = \langle u, v \rangle$

and

$\qquad \large (\int_{0}^{1}f^2(t)dt = \langle u, u \rangle = ||u||^2$

and

$\qquad \large (\int_{0}^{1}g^2(t)dt = \langle v, v \rangle = ||v||^2$

so again, by Cauchy-Schwartz Inequality the inequality holds

------------------------------

## 4C) 
Let ${W}$ be the subspace of the Euclidean space $\mathbb{R}^4$ with standard inner product with basis ${S}$ = {$u_1$, $u_2$, $u_3$}, where

$\vec{u_1}$ = $\begin{bmatrix} 1 \\ 1 \\ 1 \\ 0 \end{bmatrix} \qquad$
$\vec{u_2}$ = $\begin{bmatrix} -1 \\ 0 \\ -1 \\ 1 \end{bmatrix} \qquad$
$\vec{u_3}$ = $\begin{bmatrix} -1 \\ 0 \\ 0 \\ -1 \end{bmatrix} \qquad$

Transform ${S}$ to an orthonormal basis ${T}$ = {$w_1$, $w_2$, $w_3$} using the Gram-Schmidt process.

-----------------------

### **answer**

Define $\qquad \large w_1 = \frac{u_1}{||u_1||} = \frac{1}{\sqrt{3}} \begin{pmatrix} 1 \\ 1\\1\\0 \end{pmatrix}$

Then compute $\quad \large v_2 = u_2 - \langle u_2, w_1 \rangle w_1$

$\qquad \large =\begin{pmatrix}-1\\0\\-1\\1 \end{pmatrix} + (\frac{2}{\sqrt{3}}) \frac{1}{\sqrt{3}} \begin{pmatrix}1\\1\\1\\0\end{pmatrix}$
$\quad \Rightarrow \large \begin{pmatrix} -1 / 3 \\ 2 / 3 \\ -1 / 3 \\ 1 \end{pmatrix}$

and, 
$\qquad \large w_2 = \frac{v_2}{||v||^2} = \frac{1}{\sqrt{5/3}} \begin{pmatrix} -1/3\\2/3\\-1/3\\1 \end{pmatrix} = \frac{1}{\sqrt{15}}$ $\begin{pmatrix} -1\\2\\-1\\3 \end{pmatrix}$
$\quad \large \Rightarrow \begin{pmatrix}-4/5\\3/5\\1/5\\-3/5\end{pmatrix}$

Next compute
$\quad \large v_3 = u_3 - \langle u_3, w_1 \rangle w_1 - \langle u_3, w_2 \rangle w_2$

$\quad \large = \begin{pmatrix}-1\\0\\0\\-1 \end{pmatrix} + (\frac{1}{\sqrt{3}}) \frac{1}{\sqrt{3}} \begin{pmatrix}1\\1\\1\\0\end{pmatrix}$
$ \large + (\frac{2}{\sqrt{15}}) \frac{1}{\sqrt{15}} \begin{pmatrix}-1\\2\\-1\\3 \end{pmatrix}$
$\quad \large \Rightarrow \begin{pmatrix} -4/5\\3/5\\1/5\\-3/5 \end{pmatrix}$

and, $\quad \large w_3 = \frac{v_3}{||v_3||} = \frac{1}{sqrt{35}} \begin{pmatrix} -4\\3\\1\\-3 \end{pmatrix}$

Finally, the orthonormal basis is 

$T = \Bigg\{ \large\frac{1}{\sqrt{3}} \begin{pmatrix}1\\1\\1\\0\end{pmatrix}, \frac{1}{\sqrt{15}} \begin{pmatrix}-1\\2\\-1\\3\end{pmatrix}, \frac{1}{\sqrt{35}} \begin{pmatrix}-4\\3\\1\\-3\end{pmatrix} \Bigg\}$

## **6E)**

Which of the following matrices are positive-definite?

A = $\begin{bmatrix} 2 & 1 & 1 \\ 1 & 2 & 1 \\ 1 & 1 & 2 \end{bmatrix} \qquad$
B = $\begin{bmatrix} 3 & 2 \\ 2 & 5 \end{bmatrix} \qquad$
C = $\begin{bmatrix} 1 & 4 & 5 \\ 0 & 2 & 6 \\ 0 & 0 & 3 \end{bmatrix} \qquad$
E = $\begin{bmatrix} 1 & 3 \\ 3 & 5 \end{bmatrix} \qquad$

### answers

A symmetric matrix is positive definite if and only if all the eigenvlaues are positive

**For A,**

$\left|\begin{matrix}2-\lambda & 1 & 1 \\1 & 2-\lambda & 1 \\1 & 1 & 2-\lambda \end{matrix}\right|$
$ = \lambda^3 + 6 \times \lambda^2 -9 \times \lambda + 4$

$ = -(\lambda -1) \times (\lambda^2 -5 \times \lambda +4)$
$ = -(\lambda -1)\times(\lambda -1)\times(\lambda -4)$

The eigenvaues of $A$ are 1, 1, and 4 so $A$ is positive definite

--------------------------

**For B,**

$\left|\begin{matrix}3-\lambda & 2 \\2 & 5-\lambda\end{matrix}\right| = \lambda^2 -8 \times \lambda + 11$

$ = (\lambda + (\sqrt{5} - 4))\times(\lambda - (/sqrt{5} + 4))$

The eigenvaues of $B$ are 4 $\pm \sqrt{5}$ so $B$ is positive definite

------------------------------

**For C,**

let $v = \begin{pmatrix} x\\y\\z\end{pmatrix}$ be a nonzero vector

$v^TCv = $

$ = x^2 + 4xy +2y^2 +5xz +6yz +3z^2$

next we determine that for $v = \begin{pmatrix}0\\1\\-1\end{pmatrix}, v^TCv = -1 < 0$ so, $C$ is not positive definite

----------------

**For E,**

$\left|\begin{matrix}1-\lambda & 3 \\3 & 5-\lambda\end{matrix}\right|$
